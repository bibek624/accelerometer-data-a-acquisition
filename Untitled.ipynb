```python
import numpy as np
from scipy.signal import butter, lfilter
import matplotlib.pyplot as plt

# Generate a noisy signal
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)  # Time vector
signal = np.sin(2 * np.pi * 5 * t)  # A 5 Hz sine wave
noise = 0.5 * np.random.normal(size=len(t))  # Gaussian noise
noisy_signal = signal + noise

# Define the low-pass filter
cutoff_frequency = 5  # Cutoff frequency in Hz
nyquist_frequency = 0.5 * fs
b, a = butter(1, cutoff_frequency / nyquist_frequency, btype='low')

# Apply the filter
filtered_signal = lfilter(b, a, noisy_signal)

# Plot the original, noisy, and filtered signals
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Original Signal', lw=2)
plt.plot(t, noisy_signal, label='Noisy Signal', alpha=0.7)
plt.plot(t, filtered_signal, label='Filtered Signal', lw=2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Low-Pass Butterworth Filter Example')
plt.grid(True)
plt.show()
```
